100ページ・見開き対応：PDF→PNGをバッチ変換

```python
import os
from pdf2image import convert_from_path

# PDF_PATH = "../data/raw/purge_list_image/purge_list_image_part2_1.pdf"
PDF_PATH = "purge_list_image_part2_1.pdf"
OUT_DIR = "../../../output/purge_part2_1/spreads"
POPPLER_PATH = "/opt/homebrew/opt/poppler/bin"

DPI = 400      # まず400推奨。文字が潰れるなら600へ
BATCH = 10     # 10ページずつ（メモリ安全）
TOTAL_PAGES = 100

os.makedirs(OUT_DIR, exist_ok=True)

for start in range(1, TOTAL_PAGES + 1, BATCH):
    end = min(start + BATCH - 1, TOTAL_PAGES)
    print(f"Convert pages {start}-{end} ...")

    pages = convert_from_path(
        PDF_PATH,
        dpi=DPI,
        fmt="png",
        first_page=start,
        last_page=end,
        poppler_path=POPPLER_PATH,
        thread_count=4,
    )

    for i, page in enumerate(pages, start=start):
        out_path = os.path.join(OUT_DIR, f"spread_{i:03}.png")
        page.save(out_path, "PNG")

    del pages

print("✅ spreads done")
```
